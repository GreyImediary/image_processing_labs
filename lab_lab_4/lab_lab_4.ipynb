```python
from scipy.fft import fft, fftfreq
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
def spectr(image):
    imageSpect = np.fft.fft2(image)
    return imageSpect
```

```python
def centr(image):
    imageSpect = np.fft.fft2(image)
    fshift = np.fft.fftshift(imageSpect)
    return fshift
```

```python
def decentr(image):
    imageSpect = np.fft.fft2(image)
    fshift = np.fft.fftshift(imageSpect)
    ifshift= np.fft.ifftshift(fshift)
    return ifshift

```

```python

def process(image):
    imageSpect = np.fft.fft2(image)
    fshift = np.fft.fftshift(imageSpect)
    ifshift= np.fft.ifftshift(fshift)
    iff = np.fft.ifft2(ifshift)
    proc = (np.abs(iff))
    return proc
```

```python
fig = plt.figure()

fig, ([ax1, ax11],[ax2, ax22], [ax3, ax33], [ax4, ax44], [ax5, ax55], [ax6, ax66] ) = plt.subplots(
    nrows=6, ncols=2,
    figsize=(18, 24)
)

ax1.set_title('01')
ax2.set_title('02')
ax3.set_title('03')
ax4.set_title('04')
ax5.set_title('05')
ax6.set_title('06')

image1 = cv2.imread("4_01.jpg") # .
img1 = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

image2 = cv2.imread("4_02.jpg") # .
img2 = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)

image3 = cv2.imread("4_03.jpg") # .
img3 = cv2.cvtColor(image3,cv2.COLOR_BGR2GRAY)

image4 = cv2.imread("4_04.jpg") # .
img4 = cv2.cvtColor(image4,cv2.COLOR_BGR2GRAY)

image5 = cv2.imread("4_05.jpg") # .
img5 = cv2.cvtColor(image5,cv2.COLOR_BGR2GRAY)

image6 = cv2.imread("4_06.jpg") # .
img6 = cv2.cvtColor(image6,cv2.COLOR_BGR2GRAY)

image1S = centr(img1)
image2S = centr(img2)
image3S = centr(img3)
image4S = centr(img4)
image5S = centr(img5)
image6S = centr(img6)

ax1.imshow(image1, cmap='gray')
ax11.imshow(np.log(1+np.abs(image1S)),cmap='gray')

ax2.imshow(image2,cmap='gray')
ax22.imshow(np.log(1+np.abs(image2S)),cmap='gray')

ax3.imshow(image3, cmap='gray')
ax33.imshow(np.log(1+np.abs(image3S)),cmap='gray')

ax4.imshow(image4,cmap='gray')
ax44.imshow(np.log(1+np.abs(image4S)),cmap='gray')

ax5.imshow(image5, cmap='gray')
ax55.imshow(np.log(1+np.abs(image5S)),cmap='gray')

ax6.imshow(image6,cmap='gray')
ax66.imshow(np.log(1+np.abs(image6S)),cmap='gray')

plt.show()

cv2.imwrite("01.jpg", img1)
cv2.imwrite("02.jpg", img2)
cv2.imwrite("03.jpg", img3)
cv2.imwrite("04.jpg", img4)
cv2.imwrite("05.jpg", img5)
cv2.imwrite("06.jpg", np.log(1+np.abs(image6S)))

fig.savefig("spectres.jpg")
```
